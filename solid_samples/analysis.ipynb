```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
%matplotlib widget
```

Loading and Organizing Data

```python
t_df = pd.read_csv("mango_trichomes/compiled_spectral_data/mango_trichomes_16_scans.csv")
tp_df = pd.read_csv("trichomes_pollen/compiled_spectral_data/trichomes_pollen_scan16.csv")
wavenumber = pd.read_csv("wavenumber.csv")
```

```python
comp_fig, comp_axis = plt.subplots(1,1, figsize=(10,5))

comp_axis.plot(wavenumber, t_df.mean(axis=1), color='blue', label="Mango Trichomes")
comp_axis.plot(wavenumber, tp_df.mean(axis=1), color='orange', label="Trichomes + Pollen")

comp_axis.invert_xaxis()

comp_axis.set(title="Averaged Spectral Data of Mango Trichomes and Trichomes + Pollen Solid Samples",
              xlabel='Wavenumber cm$^{-1}$',
              ylabel='Absorbance')

comp_axis.legend()
comp_fig.tight_layout()
```
