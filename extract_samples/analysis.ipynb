```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
%matplotlib widget
```

Loading and Organizing Data

```python
t_df = pd.read_csv("t_pe/protein_extract_spectra/protein_extract_24_scans.csv")
tp_df = pd.read_csv("t+p_pe/protein_extract_spectra/protein_extract_24_scans.csv")
wavenumber = pd.read_csv("wavenumber.csv")
```

```python
comp_fig, comp_axis = plt.subplots(1,1, figsize=(10,5))

comp_axis.plot(wavenumber, t_df.mean(axis=1), color='blue', label="Mango Trichomes")
comp_axis.plot(wavenumber, tp_df.mean(axis=1), color='orange', label="Trichomes + Pollen")

comp_axis.invert_xaxis()

comp_axis.set(title="Averaged Spectral data of Mango Trichomes + Pollen and Trichomes Extract",
              xlabel='Wavenumber cm$^{-1}$',
              ylabel='Absorbance')

comp_axis.legend()
comp_fig.tight_layout()
```
